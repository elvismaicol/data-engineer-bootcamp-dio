### ETL - Transformação

```python
import pandas as pd
import pandera as pa
```

```python
path = 'C:/Temp/dio/arquivos/data-engineer/raw/'
```

```python
file_ocorrencia = 'ocorrencia.csv'
```

```python
valores_ausentes = ['**', '****', '*****', '####', '###!', 'NULL'] 
df_oc = pd.read_csv(path + file_ocorrencia, sep=';', parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
```

```python
df_oc.head()
```

```python
df_oc
```

```python
df = df_oc[['codigo_ocorrencia','codigo_ocorrencia2','ocorrencia_classificacao','ocorrencia_cidade','ocorrencia_uf','ocorrencia_aerodromo','ocorrencia_dia','ocorrencia_hora','total_recomendacoes']]
```

```python
df.head()
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "col_ficticia_codigo" : pa.Column(pa.Int, required=False),
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String), 
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable=True), 
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```

```python
# obter ocorrencias com mais de 10 recomentações
filtro = df.total_recomendacoes > 10
df.loc[filtro]
```

```python
# obter os dados de uma coluna filtradoss pelos dados de uma outra coluna(ex.: total_recomendacoes) 
filtro = df.total_recomendacoes > 10
df.loc[filtro, 'ocorrencia_cidade']
```

```python
# obter os dados de várias colunas, filtrados pelos dados de uma outra coluna(ex.: total_recomendacoes) 
filtro = df.total_recomendacoes > 10
df.loc[filtro, ['ocorrencia_cidade', 'total_recomendacoes']]
```

```python
# filtrando os dados de uma coluna usando um critério
filtro = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
df.loc[filtro, 'ocorrencia_classificacao']
```

```python
# filtrando os dados do DF usando uma coluna com um critério
filtro = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
df.loc[filtro]
```

```python
# filtrando os dados do DF usando mais de 1 coluna e critérios
# & (operador E)
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
# filtrando os dados do DF usando mais de 1 coluna e critérios
# | (operador OU)
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 | filtro2]
```

```python
# filtrando os dados do DF usando mais de 1 coluna e critérios
# | (operador OU) e & (operador E)
filtro1 = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE')
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
# filtrando os dados do DF usando mais de 1 coluna e critérios
# isin() -> verifica se cada elemento no DataFrame está nos valores fornecidos
filtro1 = df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE', 'INCIDENTE'])
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```
